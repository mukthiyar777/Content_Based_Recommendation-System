```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from ast import literal_eval
```

```python
df=pd.read_csv("C:\\Users\\Venky\\Desktop\\tmdb_5000_movies.csv")
df.info()
```

```python
df['genres']=df['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance (x,list) else [])
```

```python
df.head()
```

```python
links_small=pd.read_csv("C:\\Users\\Venky\\Desktop\\links_small.csv")
links_small=links_small[links_small['tmdbId'].notnull()]['tmdbId'].astype('int')
```

```python
df['id']=df['id'].astype('int')
df=df[df['id'].isin(links_small)]
df.head(2)
```

```python
df['tagline']=df['tagline'].fillna('')
df['description']=df['overview']+df['tagline']
df['description']=df['description'].fillna('')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(analyzer='word',ngram_range=(1,2),min_df=0,stop_words='english')
tfidf_matrix=tf.fit_transform(df['description'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cos_similarity=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
df=df.reset_index()
```

```python
df.head()
```

```python
titles=df['title']
indices=pd.Series(df.index, index=df['title'])
```

```python
indices.shape
```

```python
cos_similarity.shape
```

```python
def content_based_recommendations(title):
    idx=indices[title]
    cos_sim_scores=list(enumerate(cos_similarity[idx]))
    cos_sim_scores=sorted(cos_sim_scores,key=lambda x: x[1],reverse=True)
    cos_sim_scores=cos_sim_scores[1:31]
    movie_indices=[i[0] for i in cos_sim_scores]
    return titles.iloc[movie_indices]
```
